```python
import matplotlib.pyplot as plt
import numpy as np

def real_func(t):
    return np.exp(-t) + t - 1

real_func_space = np.linspace(0, 0.4, 50)
real_func_values = real_func(real_func_space)

```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')

#plt.style.use('dark_background')
#Plot a line graph

fig = plt.figure(figsize=(10, 10))
plt.plot(real_func_space, real_func_values, label = 'Analitik Çözüm', alpha = 0.5 , )
plt.plot(np.linspace(0, 0.4, 5),[0, 0.005, 0.0190, 0.0411, 0.0707], label='Heun',linestyle = "--", color="blue" )
plt.plot(np.linspace(0, 0.4, 5),[0, 0.0048, 0.0187, 0.0407,0.0702], label='RK-4',linestyle = "--", color = "red")

 
# Add labels and title
plt.title("Analitik Çözüm & Heun & RK-4 Grafiği")
plt.xlabel("ti değerleri")
plt.ylabel("yi değerleri")
    
plt.legend() 

plt.show()
fig.savefig('all')
```

```python
#Plot a line graph
plt.style.use('dark_background')
fig = plt.figure(figsize=(10, 10))
plt.plot(real_func_space, real_func_values, label = 'Analitik Çözüm', alpha = 0.5)
plt.plot(np.linspace(0, 0.4, 5),[0, 0.005, 0.0190, 0.0411, 0.0707], label='Heun Metodu', linestyle = '--', color = "red" )

for t, sol in zip(np.linspace(0, 0.4, 5), [0, 0.005, 0.0190, 0.0411, 0.0707]):
    plt.plot(t, sol, marker='.', markersize=15, markeredgecolor="k", markeredgewidth=1, label = 't = {}'.format(np.round(t, 2)), color = 'blue')

 
# Add labels and title
plt.title("Analitik Çözüm & Heun Metodu ")
plt.xlabel("ti değerleri")
plt.ylabel("Yi değerleri")
 
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
plt.style.use('dark_background')
#Plot a line graph
plt.plot(real_func_space, real_func_values, label = 'Analitik Çözüm', alpha = 0.5)
plt.plot(np.linspace(0, 0.4, 5),[0, 0.0048, 0.0187, 0.0407,0.0702], label='Runge-Kutta-4',linestyle = '--', color = "red")

for t, sol in zip(np.linspace(0, 0.4, 5), [0, 0.0048, 0.0187, 0.0407,0.0702]):
    plt.plot(t, sol, marker='.', markersize=15, markeredgecolor="k", markeredgewidth=1, label = 't = {}'.format(np.round(t, 2)), color = 'blue')

# Add labels and title
plt.title("Analitik Çözüm & Runge-Kutta-4")
plt.xlabel("ti değerleri")
plt.ylabel("yi değerleri")
 
plt.legend()
plt.show()
```

```python
def real_func_euler(t):
    return np.exp(-2*t)*np.sin(3*t) + 3*np.cos(t)+np.sin(t)


real_func_space_euler = np.linspace(0, 1, 50)
real_func_values_euler = real_func_euler(real_func_space_euler)
```

```python
plt.style.use('seaborn')
plt.style.use('dark_background')
fig = plt.figure(figsize=(10, 10))

#Plot a line graph
plt.plot(real_func_space_euler, real_func_values_euler, label = 'Analitik Çözüm', alpha = 0.5)
plt.plot(np.linspace(0, 1, 6),[3, 3.8, 4, 3.6321,2.9522,2.2480], label='Euler Sayısal Çözümü',linestyle = '--',color = "yellow" )

 
# Add labels and title
plt.title("Analitik Çözüm & Euler Yöntemi")
plt.xlabel("ti değerleri")
plt.ylabel("xi değerleri")
 
plt.legend()
plt.show()
```
